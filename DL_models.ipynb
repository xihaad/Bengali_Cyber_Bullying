```python
from bnlp import NLTKTokenizer
import pandas as pd

# # Example data
# data = {
#     "Description": [
#         "আমি ভাত খাই। সে বাজারে যায়। তিনি কি সত্যিই ভালো মানুষ?",
#         # ... more Bengali text descriptions ...
#     ],
#     "Label": ["neutral", "sexual", "troll", "threat", "political", # ... corresponding labels ... ]
# }

# df = pd.DataFrame(data)


bnltk = NLTKTokenizer()

def preprocess_text(text):
    # Tokenize words and sentences
    word_tokens = bnltk.word_tokenize(text)
    sentence_tokens = bnltk.sentence_tokenize(text)


    return {
        "word_tokens": word_tokens,
        "sentence_tokens": sentence_tokens
    }

# Apply the preprocessing function to the entire dataset
df["processed_text"] = df["Description"].apply(preprocess_text)

# Display the processed DataFrame
print(df)
```

**Text Cleaning**

```python
from bnlp import CleanText

clean_text = CleanText(
   fix_unicode=True,
   unicode_norm=True,
   unicode_norm_form="NFKC",
   remove_url=False,
   remove_email=False,
   remove_emoji=False,
   remove_number=False,
   remove_digits=False,
   remove_punct=False,
   replace_with_url="<URL>",
   replace_with_email="<EMAIL>",
   replace_with_number="<NUMBER>",
   replace_with_digit="<DIGIT>",
   replace_with_punct = "<PUNC>"
)

df['clean_text'] = clean_text(df["processed_text"])
df['clean_descripted_text'] = clean_text(df["Description"])
print(clean_text)
```

```python


import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, GRU, LSTM, Dense
from bnlp import BasicTokenizer



def tokenize_bengali_text(text):
    tokenizer = BasicTokenizer()
    tokens = tokenizer.tokenize(text)
    return tokens

max_text_length = 200
num_words = 10000
df['Tokenized_Description'] = df['Description'].apply(tokenize_bengali_text)


tokenizer = Tokenizer(num_words=num_words, oov_token='<OOV>')
tokenizer.fit_on_texts(df['Tokenized_Description'])
word_index = tokenizer.word_index

sequences = tokenizer.texts_to_sequences(df['Tokenized_Description'])
padded_sequences = pad_sequences(sequences, maxlen=max_text_length, padding='post', truncating='post')


labels = pd.get_dummies(df['Label']).values

```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(padded_sequences, labels, test_size=0.2, random_state=42)

# Define the models
def build_rnn_model():
    model = Sequential([
        Embedding(input_dim=num_words, output_dim=64),
        SimpleRNN(64, return_sequences=True),
        SimpleRNN(32),
        Dense(16, activation='relu'),
        Dense(5, activation='softmax')
    ])
    return model


rnn_model = build_rnn_model()
rnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
rnn_model.fit(X_train, y_train, epochs=30, validation_data=(X_test, y_test))


```

```python
rnn_model.summary()
```

```python
def build_gru_model():
    model = Sequential([
        Embedding(input_dim=num_words, output_dim=64),
        GRU(64, return_sequences=True),
        GRU(32),
        Dense(16, activation='relu'),
        Dense(5, activation='softmax')
    ])
    return model



gru_model = build_gru_model()
gru_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
gru_model.fit(X_train, y_train, epochs=30, validation_data=(X_test, y_test))

```

```python
gru_model.summary()
```

```python
def build_lstm_model():
    model = Sequential([
        Embedding(input_dim=num_words, output_dim=64),
        LSTM(64, return_sequences=True),
        LSTM(32),
        Dense(16, activation='relu'),
        Dense(8, activation='relu'),
        Dense(5, activation='softmax')
    ])
    return model

lstm_model = build_lstm_model()
lstm_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
lstm_model.fit(X_train, y_train, epochs=30, validation_data=(X_test, y_test))
```
