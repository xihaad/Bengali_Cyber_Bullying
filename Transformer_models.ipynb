```python
from bnlp import NLTKTokenizer
import pandas as pd



bnltk = NLTKTokenizer()

def preprocess_text(text):

    word_tokens = bnltk.word_tokenize(text)
    sentence_tokens = bnltk.sentence_tokenize(text)


    return {
        "word_tokens": word_tokens,
        "sentence_tokens": sentence_tokens
    }


df["processed_text"] = df["Description"].apply(preprocess_text)


print(df)
```

**Text Cleaning**

```python
from bnlp import CleanText

clean_text = CleanText(
   fix_unicode=True,
   unicode_norm=True,
   unicode_norm_form="NFKC",
   remove_url=False,
   remove_email=False,
   remove_emoji=False,
   remove_number=False,
   remove_digits=False,
   remove_punct=False,
   replace_with_url="<URL>",
   replace_with_email="<EMAIL>",
   replace_with_number="<NUMBER>",
   replace_with_digit="<DIGIT>",
   replace_with_punct = "<PUNC>"
)

df['clean_text'] = clean_text(df["processed_text"])
df['clean_descripted_text'] = clean_text(df["Description"])
print(clean_text)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, GRU, LSTM, Dense, Bidirectional, LSTM, GRU, Input
from tensorflow.keras.models import Model
from tensorflow.keras.layers import MultiHeadAttention
from bnlp import BasicTokenizer, BengaliWord2Vec



def tokenize_bengali_text(text):
    tokenizer = BasicTokenizer()
    tokens = tokenizer.tokenize(text)
    return tokens


bwv = BengaliWord2Vec()
model_path = "/content/bnwiki_word2vec.model"

max_text_length = 200
num_words = 100000


def text_to_vector(text):
    tokens = tokenize_bengali_text(text)
    print(tokens)
    vector = np.zeros((max_text_length, 100))
    # print(vector)
    for i, token in enumerate(tokens[:max_text_length]):
        if token in bwv.model.wv.key_to_index:
            vector[i, :] = bwv.model.wv.get_vector(token)
            # print(vector)
    return vector



vectors = np.array([text_to_vector(text) for text in df['Description']])
labels = pd.get_dummies(df['Label']).values


```

```python
X_train, X_test, y_train, y_test = train_test_split(vectors, labels, test_size=0.2, random_state=42)


def build_transformer_model():
    inputs = Input(shape=(max_text_length, 100))
    x = MultiHeadAttention(num_heads=6, key_dim=300)(inputs, inputs)
    x = Bidirectional(LSTM(64, return_sequences=True))(x)
    x = Bidirectional(LSTM(32))(x)
    x = Dense(16, activation='relu')(x)
    outputs = Dense(5, activation='softmax')(x)

    model = Model(inputs=inputs, outputs=outputs)
    return model


transformer_model = build_transformer_model()
transformer_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
transformer_model.fit(X_train, y_train, epochs=60, validation_data=(X_test, y_test))
```
